Part 1: Analyzing Best-Seller Trends Across Product Categories

**Objective**: Understand the relationship between product categories and their best-seller status.

1. **Crosstab Analysis**:
    - Create a crosstab between the product `category` and the `isBestSeller` status.
    
    - Are there categories where being a best-seller is more prevalent? 
    	
    	*Hint: one option is to calculate the proportion of best-sellers for each category and then sort the categories based on this proportion in descending order.*


2. **Statistical Tests**:
    - Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
    - Compute Cramér's V to understand the strength of association between best-seller status and category.

3. **Visualizations**:
	- Visualize the relationship between product categories and the best-seller status using a stacked bar chart.

```python
# Import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis
```

```python
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
#Create crosstab
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])

crosstab_result
```

```python
#Calculate the total count for each category (best-sellers + non-best-sellers)
# Directly referencing the columns in crosstab_result
crosstab_result['Total'] = crosstab_result[False] + crosstab_result[True]

# Calculate the proportion of best-sellers
crosstab_result['Proportion_BestSeller'] = crosstab_result[True] / crosstab_result['Total']

#Sort the categories by the proportion of best-sellers in descending order
sorted_proportions = crosstab_result.sort_values(by='Proportion_BestSeller', ascending=False)
sorted_proportions
# A higher value in Proportion_BestSeller indicates a greater proportion of products in that category are best-sellers.
```

```python
from scipy.stats import chi2_contingency

# Chi-square test for 'MSZoning' and 'SaleCondition'
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result)

chi2_statistic, chi2_p_value
```

High Chi-Square Statistic: Indicates a significant difference between observed and expected counts, suggesting a strong relationship between the variables.
p-value of 0.0: Indicates that this relationship is statistically significant, and it is highly unlikely that the observed association happened by chance.

```python
from scipy.stats.contingency import association

# Computing the association between variables in 'crosstab_result' using the "cramer" method
association(crosstab_result, method="cramer")
```

The Cramér's V value of 0.1222 suggests a weak association between the two variables. Though statistically significant, the strength of this relationship is not very strong in practical terms.

```python
# Plotting a stacked bar chart for the 'crosstab_result' data
sorted_proportions.head(5).plot(kind="bar", stacked=True)
```

### Part 2: Exploring Product Prices and Ratings Across Categories and Brands

**Objective**: Investigate how different product categories influence product prices.

0. **Preliminary Step: Remove outliers in product prices.**

	For this purpose, we can use the IQR (Interquartile Range) method. Products priced below the first quartile minus 1.5 times the IQR or above the third quartile plus 1.5 times the IQR will be considered outliers and removed from the dataset. The next steps will be done with the dataframe without outliers.
	
	*Hint: you can check the last Check For Understanding at the end of the lesson EDA Bivariate Analysis for a hint on how to do this.*

1. **Violin Plots**:
    - Use a violin plot to visualize the distribution of `price` across different product `categories`. Filter out the top 20 categories based on count for better visualization.
    - Which product category tends to have the highest median price? Don't filter here by top categories.

2. **Bar Charts**:
    - Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
    - Which product category commands the highest average price? Don't filter here by top categories.

3. **Box Plots**:
    - Visualize the distribution of product `ratings` based on their `category` using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.
    - Which category tends to receive the highest median rating from customers? Don't filter here by top categories.

```python
def tukeys_test_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds for the outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify the outliers
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    return outliers
```

```python
# Apply Tukey's test on the 'price' column
df['price_no_outliers'] = tukeys_test_outliers(df['price'])

# Filter the top 20 categories based on the count of products (after removing outliers)
top_20_categories = df.groupby('category')['price_no_outliers'].count().nlargest(20).index

# Filter the DataFrame to include only the top 20 categories
df_top_20 = df[df['category'].isin(top_20_categories)]
```

```python
df_top_20
```

```python
# Plotting a violin plot to visualize the distribution of 'SalePrice' for each 'MSZoning' category, using the "coolwarm" color palette
sns.violinplot(data=df_top_20, x='category', y='price_no_outliers', palette="coolwarm")
```

```python
# Filter the top 10 categories based on the count of products (after removing outliers)
top_10_categories = df.groupby('category')['price_no_outliers'].count().nlargest(20).index

# Filter the DataFrame to include only the top 10 categories
df_top_10 = df[df['category'].isin(top_10_categories)]

# Calculate the average price for each category in the top 10
avg_price_by_category = df_top_10.groupby('category')['price_no_outliers'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_by_category.index, y=avg_price_by_category.values, palette="Set2")
plt.xticks(rotation=90)  # Rotate the x-axis labels for readability
plt.title('Average Price of Products Across Top 10 Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Average Price')
plt.show()
```

```python
# Filter the top 10 categories based on the count of products
top_10_categories = df.groupby('category')['stars'].count().nlargest(10).index

# Filter the DataFrame to include only the top 10 categories
df_top_10 = df[df['category'].isin(top_10_categories)]

# Create side-by-side box plots for the rating distribution across the top 10 categories
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='stars', data=df_top_10, palette="Set3")

plt.xticks(rotation=90)  # Rotate the x-axis labels for readability
plt.title('Product Rating Distribution Across Top 10 Categories')
plt.xlabel('Product Category')
plt.ylabel('Rating')
plt.show()
```

```python
# Calculate the median rating for each category
median_ratings_by_category = df.groupby('category')['stars'].median()

# Find the category with the highest median rating
highest_median_category = median_ratings_by_category.idxmax()

# Filter the DataFrame to include only the category with the highest median rating
df_highest_median = df[df['category'] == highest_median_category]

df_highest_median
```

### Part 3: Investigating the Interplay Between Product Prices and Ratings

**Objective**: Analyze how product ratings (`stars`) correlate with product prices.

1. **Correlation Coefficients**:
    - Calculate the correlation coefficient between `price` and `stars`.
    - Is there a significant correlation between product price and its rating?
	
2. **Visualizations**:
    - Use a scatter plot to visualize the relationship between product rating and price. What patterns can you observe?
    - Use a correlation heatmap to visualize correlations between all numerical variables.
    - Examine if product prices typically follow a normal distribution using a QQ plot. 


```python
# Calculating the Pearson correlation coefficient between 'price_no_outliers' and 'stars columns
correlation = df['price_no_outliers'].corr(df['stars'])
correlation
```

```python
# Calculating the Spearman rank correlation between 'price_no_outliers' and 'stars' 
correlation = df['price_no_outliers'].corr(df['stars'], method='spearman')
correlation
```

The variables show a weak inverse relationship, meaning as one increases, the other tends to decrease slightly.
The relationship captured by Spearman suggests a more consistent trend (not necessarily linear), whereas Pearson is focused on linearity and shows a slightly weaker relationship.
In practice, these weak correlations suggest that while there is a slight tendency for the variables to move in opposite directions, it may not be significant enough for predictive purposes or strong conclusions
